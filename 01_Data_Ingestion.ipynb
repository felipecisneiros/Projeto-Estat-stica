```python
# Criar o schema bronze no catálogo projeto_estatistica
spark.sql("CREATE SCHEMA IF NOT EXISTS projeto_estatistica.bronze")
print("✅ Schema bronze criado com sucesso!")
```

```python
spark.sql("USE CATALOG projeto_estatistica")
spark.sql("USE SCHEMA bronze")  

tabelas = [
    "dim_customer",
    "dim_delivery", 
    "dim_products",
    "dim_shopping",
    "fact_orders"
]

for tabela in tabelas:
    print(f"Lendo tabela {tabela}...")

    df = spark.table(f"projeto_estatistica.default.{tabela}")

    # Salvar diretamente como tabela Delta no schema bronze
    df.write.format("delta").mode("overwrite").saveAsTable(f"bronze.{tabela}")

    print(f"✔ Tabela {tabela} salva como projeto_estatistica.bronze.{tabela}\n")
```
